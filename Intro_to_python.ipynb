```python
def year_range(x,y):
    list_dict = {}
    for year in range(x, y+1):
        year_str = 'Y' + str(year)
        list_dict[year_str] = ['sum']

    return(list_dict)
```

```python
year_range(2014,2018)
```

```python
element_grp.agg(year_range(2014,2018)).loc['Stock Variation',:]
```
